```python
def predict_match_output(home_team, away_team, clf_xg, clf_goals, handicap=-0.50):#ใส่ราคา
    probs_xg = clf_xg.predict(home_team, away_team)
    probs_goals = clf_goals.predict(home_team, away_team)

    # Averages between the two models
    avg_home_goal_expectation = (probs_xg.home_goal_expectation + probs_goals.home_goal_expectation) / 2
    avg_away_goal_expectation = (probs_xg.away_goal_expectation + probs_goals.away_goal_expectation) / 2
    avg_home_win = (probs_xg.home_win + probs_goals.home_win) / 2
    avg_draw = (probs_xg.draw + probs_goals.draw) / 2
    avg_away_win = (probs_xg.away_win + probs_goals.away_win) / 2
    avg_both_teams_to_score = (probs_xg.both_teams_to_score + probs_goals.both_teams_to_score) / 2

    # Calculate the probability of there being over 1.5 goals in the match
    over_1_5_goals_probability_xg = probs_xg.total_goals("over", -3.25)
    over_1_5_goals_probability_goals = probs_goals.total_goals("over", -3.25)
    avg_over_1_5_goals_probability = (over_1_5_goals_probability_xg + over_1_5_goals_probability_goals) / 2

    # Predicted Score based on rounding off
    predicted_score_xg = (round(probs_xg.home_goal_expectation), round(probs_xg.away_goal_expectation))
    predicted_score_goals = (round(probs_goals.home_goal_expectation), round(probs_goals.away_goal_expectation))

    # Asian Handicap
    home_win_with_handicap_xg = probs_xg.asian_handicap("home", handicap)
    away_win_with_handicap_xg = 1 - home_win_with_handicap_xg
    home_win_with_handicap_goals = probs_goals.asian_handicap("home", handicap)
    away_win_with_handicap_goals = 1 - home_win_with_handicap_goals

    # Display results
    print(f"Predictions based on Expected Goals (xG):")
    print(f"Home Goal Expectation: {probs_xg.home_goal_expectation:.2f}")
    print(f"Away Goal Expectation: {probs_xg.away_goal_expectation:.2f}")
    print(f"Predicted Score: {home_team} {predicted_score_xg[0]} - {away_team} {predicted_score_xg[1]}")
    print(f"Probability of {home_team} win: {probs_xg.home_win*100:.2f}%")
    print(f"Probability of draw: {probs_xg.draw*100:.2f}%")
    print(f"Probability of {away_team} win: {probs_xg.away_win*100:.2f}%")
    print(f"Probability of Both Teams to Score: {probs_xg.both_teams_to_score*100:.2f}%")
    print(f"With {handicap} handicap:")
    print(f"Probability of {home_team} win: {home_win_with_handicap_xg*100:.2f}%")
    print(f"Probability of {away_team} win: {away_win_with_handicap_xg*100:.2f}%")
    print(f"Probability of Over 1.5 Goals: {over_1_5_goals_probability_xg*100:.2f}%")
    print("\n" + "-" * 60 + "\n")
    print(f"Predictions based on Actual Goals:")
    print(f"Home Goal Expectation: {probs_goals.home_goal_expectation:.2f}")
    print(f"Away Goal Expectation: {probs_goals.away_goal_expectation:.2f}")
    print(f"Predicted Score: {home_team} {predicted_score_goals[0]} - {away_team} {predicted_score_goals[1]}")
    print(f"Probability of {home_team} win: {probs_goals.home_win*100:.2f}%")
    print(f"Probability of draw: {probs_goals.draw*100:.2f}%")
    print(f"Probability of {away_team} win: {probs_goals.away_win*100:.2f}%")
    print(f"Probability of Both Teams to Score: {probs_goals.both_teams_to_score*100:.2f}%")
    print(f"With {handicap} handicap:")
    print(f"Probability of {home_team} win: {home_win_with_handicap_goals*100:.2f}%")
    print(f"Probability of {away_team} win: {away_win_with_handicap_goals*100:.2f}%")
    print(f"Probability of Over 1.5 Goals: {over_1_5_goals_probability_goals*100:.2f}%")

    print("\n" + "-" * 60 + "\n")
    print(f"Average Probability of Over 1.5 Goals: {avg_over_1_5_goals_probability*100:.2f}%")
```

```python
def combined_predict_match(home_team, away_team, clf_xg, clf_goals, handicap=-1.25):
    probs_xg = clf_xg.predict(home_team, away_team)
    probs_goals = clf_goals.predict(home_team, away_team)

    # Averages between the two models
    avg_home_goal_expectation = (probs_xg.home_goal_expectation + probs_goals.home_goal_expectation) / 2
    avg_away_goal_expectation = (probs_xg.away_goal_expectation + probs_goals.away_goal_expectation) / 2
    avg_home_win = (probs_xg.home_win + probs_goals.home_win) / 2
    avg_draw = (probs_xg.draw + probs_goals.draw) / 2
    avg_away_win = (probs_xg.away_win + probs_goals.away_win) / 2
    avg_both_teams_to_score = (probs_xg.both_teams_to_score + probs_goals.both_teams_to_score) / 2

    # Probability of Over 1.5 Goals
    over_1_5_goals_prob_xg = probs_xg.total_goals("over", 1.5)
    over_1_5_goals_prob_goals = probs_goals.total_goals("over", 1.5)
    avg_over_1_5_goals_prob = (over_1_5_goals_prob_xg + over_1_5_goals_prob_goals) / 2

    # Probability of Under 2.5 Goals
    under_2_5_goals_prob_xg = probs_xg.total_goals("under", 2.5)
    under_2_5_goals_prob_goals = probs_goals.total_goals("under", 2.5)
    avg_under_2_5_goals_prob = (under_2_5_goals_prob_xg + under_2_5_goals_prob_goals) / 2

    # Asian Handicap
    home_win_with_handicap_xg = probs_xg.asian_handicap("home", handicap)
    away_win_with_handicap_xg = 1 - home_win_with_handicap_xg
    home_win_with_handicap_goals = probs_goals.asian_handicap("home", handicap)
    away_win_with_handicap_goals = 1 - home_win_with_handicap_goals
    avg_home_win_with_handicap = (home_win_with_handicap_xg + home_win_with_handicap_goals) / 2
    avg_away_win_with_handicap = (away_win_with_handicap_xg + away_win_with_handicap_goals) / 2

    # Display results
    predicted_score_home = round(avg_home_goal_expectation)
    predicted_score_away = round(avg_away_goal_expectation)

    print(f"\nPredicted Score: {home_team} {predicted_score_home} - {away_team} {predicted_score_away}")
    print(f"Home Goal Expectation: {avg_home_goal_expectation:.2f}")
    print(f"Away Goal Expectation: {avg_away_goal_expectation:.2f}")
    print(f"Probability of {home_team} win: {avg_home_win*100:.2f}%")
    print(f"Probability of draw: {avg_draw*100:.2f}%")
    print(f"Probability of {away_team} win: {avg_away_win*100:.2f}%")
    print(f"Probability of Both Teams to Score: {avg_both_teams_to_score*100:.2f}%")
    print(f"Probability of Over 1.5 Goals: {avg_over_1_5_goals_prob*100:.2f}%")
    print(f"Probability of Under 2.5 Goals: {avg_under_2_5_goals_prob*100:.2f}%")
    print(f"With {handicap} handicap:")
    print(f"Probability of {home_team} win: {avg_home_win_with_handicap*100:.2f}%")
    print(f"Probability of {away_team} win: {avg_away_win_with_handicap*100:.2f}%")
```
